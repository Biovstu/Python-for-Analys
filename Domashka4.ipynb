### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6, 4))

sns.boxplot(data=df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('')
plt.ylabel('Стоимость')
plt.show()
```

Разброс данных очень большой.
<p>
Для наглядности срежем вылеты

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df['price'][df['price'] <= (df['price'].mean() + df['price'].std() * 1.5)])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('')
plt.ylabel('Стоимость')
plt.show()
```

```python
f"Стоимость недвижимости коллеблется от {df['price'].min()} до {df['price'].max()}"
```

```python
f"При этом среднее значение равно {df['price'].mean()}, наиболее частое {df['price'].median()}"
```

Рассмотрим изменение стоимости продаж во времени

```python
df['year'] = df['date'].apply(lambda x: int(x[:4]))
df['month'] = df['date'].apply(lambda x: int(x[4:6]))
df['month'] = df['month'] + 12 * (df['year'] - df['year'].min())
df.head()
```

```python
data = df.groupby('month')['price'].agg(min='min', mean='mean', max='max').reset_index().sort_values(by='month')
plt.figure(figsize=(6, 4))

sns.lineplot(x=data['month'], y=data['mean'])

plt.title('Динамика стоимость недвижимости')
plt.xlabel('Месяц')
plt.ylabel('Средняя стоимость')
plt.show()
```

Из графика видно, что в течение первого года средняя стоимость снижается примерно на 10%, достигая минимума в феврале 2015 года.
<p>
Затем снова поднимается.
<p>
Это скорее всего связано с сезонностью спроса - нужны данные за больший период.

```python
plt.figure(figsize=(6, 4))

sns.lineplot(x=data['month'], y=data['min'])

plt.title('Динамика стоимость недвижимости')
plt.xlabel('Месяц')
plt.ylabel('Минимальная стоимость')
plt.show()
```

Минимальная стоимость имеет более стабильно распределение, но минимум отмечается в том же феврале 2015 года, что подтверждает спад спроса.

## Задача 3
<p>
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Для начала определим одновременное влияние на цену двух признаков "береговая линия" и "вид"

```python
data_view = pd.crosstab(index = df['view'], columns = df['waterfront'], values = df['price'], aggfunc = 'mean').reset_index().sort_values(by='view', ascending=False)
data_view.rename(columns={0: 'no', 1: 'yes'}, inplace=True)
data_view
```

```python
plt.figure(figsize=(8, 6))
n_ticks = np.arange(len(data_view['no']))
w = 0.5
plt.bar(n_ticks, data_view['yes'], width=w * 1.5)
plt.bar(n_ticks, data_view['no'], width=w)


plt.title('Зависимость цены от вида и набережной')
plt.xlabel('Вид')
plt.ylabel('Средняя цена')
plt.legend(['yes', 'no'])
plt.xticks(n_ticks, data_view['view'])
plt.show()
```

Можно сделать вывод, что для домов вне набережной имеется прямая линейная зависимость средней стоимости от вида.
<p>
Для домов с набережной так же наблюдается прямая линейная зависимость средней стоимости от вида, но имеется вылет значений при виде 2.
<p>Возможно это связано с недостаточным объемом выборки (их всего 0,8%)

Теперь определим одновременное влияние возраста и наличия ремонта на среднюю стоимость домов

```python
df['renovated'] = 0
df.loc[df['yr_renovated'] > 0, ['renovated']] = 1
df['age'] = 2023 - df['yr_renovated']
df.loc[df['yr_renovated'] == 0, ['age']] = 2023 - df['yr_built']
data_age = pd.crosstab(index = df['age'], columns = df['renovated'], values = df['price'], aggfunc = 'mean').reset_index().sort_values(by='age', ascending=True)
data_age.rename(columns={0: 'no', 1: 'yes'}, inplace=True)
data_age
```

```python
plt.figure(figsize=(10, 4))
sns.lineplot(x=data_age['age'], y=data_age['no'], label='no')
sns.lineplot(x=data_age['age'], y=data_age['yes'], label='yes')
plt.title('Динамика стоимость недвижимости')
plt.xlabel('Год')
plt.ylabel('Средняя стоимость')
plt.legend()
plt.show()
```

<p>Для домов без ремонта имеется незначительная параболическая зависимость.
<p>То есть средняя стоимость выше у ссамых новых и самых старых домов, что скорее всего связано с исторической значимостью последних.
Для отремонтированных домов просматривается обратная линейная зависимость.
<p>
Проверим это.

```python
df_cut = df[df['renovated'] == 1]
sns.jointplot(x=df_cut['age'], y=df_cut['price'], kind='reg');
```

Зависимость слабая, но имеется

Для определения влияния показателей объектовов жилого фонда на цену, построим тепловую карту матрицы корреляции

```python
df.drop(columns=['id', 'date', 'zipcode', 'year', 'month', 'sqft_living15', 'sqft_lot15'], inplace=True)
```

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Как мы видим, наибольшую корреляцию цена объекта имеет с такими показателями как:
<ul>
  <li>площядь жилых помещений (sqft_living)</li>
  <li>класс (grade)</li>
  <li>площадь мансарды (sqft_above)</li>
</ul>
Изучим их подробнее

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

```python
sns.jointplot(x=df['sqft_above'], y=df['price'], kind='reg');
```

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg');
```

Все три распределения имеют коническую форму.
<p>То есть значениях опказателя меньше среднего наблюдается линейная зависимость.
<p>Но при более высоких значениях энтропия возрастает.
<p>Это скорее всего связано с тем, что на стоимость начинают оказывать одновременное воздействие другие показатели.
